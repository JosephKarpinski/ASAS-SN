```python
lrs_list = df_var['obsid'][0:10].to_list()
lrs_list
```

```python
plt.close("all")

plt.style.use('dark_background')

for idx in range(0, len(lrs_list)):
    obsid_id_target = lrs_list[idx]
    des = df_lamost_LRS_stars[df_lamost_LRS_stars['obsid'].astype(str) == str(obsid_id_target)]
    source_id_target = des['gaia_source_id'].iloc[0]
    lamost_LRS_designation = des['designation'].iloc[0]
    lamost_LRS_subclass = des['subclass'].iloc[0]
    lamost_LRS_obsid = des['obsid'].iloc[0]
    lamost_LRS_source_id = des['gaia_source_id'].iloc[0]

    # ====================================================================================
    print("\nlamost_LRS_designation: "  + str(lamost_LRS_designation))
    print("lamost_LRS_obsid: "  + str(lamost_LRS_obsid))
    print("lamost_LRS_subclass: " + str(lamost_LRS_subclass))
    print("http://www.lamost.org/dr8/v2.0/spectrum/view?obsid="  + str(lamost_LRS_obsid))

    if lamost_LRS_subclass == 'Carbon':
        print('https://en.wikipedia.org/wiki/Carbon_star')

    if lamost_LRS_subclass == 'CV':
        print('https://en.wikipedia.org/wiki/Cataclysmic_variable_star')

    if lamost_LRS_subclass == 'O':
        print('https://en.wikipedia.org/wiki/O-type_star')

    if lamost_LRS_subclass == 'OB':
        print('https://en.wikipedia.org/wiki/OB_star')

    if lamost_LRS_subclass == 'WD':
        print('https://en.wikipedia.org/wiki/White_dwarf')

    # ====================================================================================
    df_lamost_LRS_CSV = pd.read_csv('http://www.lamost.org/dr8/v2.0/spectrum/fits2csv/' + str(lamost_LRS_obsid))
    df_lamost_LRS_CSV.plot(x='Wavelength',y='Flux', 
        title = ('Gaia Source ID: ' + str(source_id_target)  + '\nlamost_LRS_obsid: ' + str(lamost_LRS_obsid) + "\nlamost_LRS_subclass: " + str(lamost_LRS_subclass)), 
        figsize=(15, 7))
    plt.annotate("H Epsilon 3970", xy=(3970, df_lamost_LRS_CSV['Flux'].median()), rotation=-90)
    plt.annotate("H Delta 4101", xy=(4101, df_lamost_LRS_CSV['Flux'].median()), rotation=-90)
    plt.annotate("H Gamma 4340", xy=(4340, df_lamost_LRS_CSV['Flux'].median()), rotation=-90)
    plt.annotate("H Beta 4861", xy=(4861, df_lamost_LRS_CSV['Flux'].median()), rotation=-90)
    plt.annotate("Mg 5176", xy=(5176, df_lamost_LRS_CSV['Flux'].median()), rotation=-90)
    plt.annotate("OI 6301", xy=(6301, df_lamost_LRS_CSV['Flux'].median()), rotation=-90)
    plt.annotate("H Alpha 6500", xy=(6500, df_lamost_LRS_CSV['Flux'].median()), rotation=-90)
    
    # ====================================================================================
    xlim = (3800, 4600)
    df_lamost_LRS_CSV_4600 = df_lamost_LRS_CSV[df_lamost_LRS_CSV['Wavelength'] <= 4600]
    df_lamost_LRS_CSV_4600.plot(x='Wavelength',y='Flux', marker='.', linestyle='--', color='b', 
        title = ('Gaia Source ID: ' + str(source_id_target)  + '\nlamost_LRS_obsid: ' + str(lamost_LRS_obsid) + "\nlamost_LRS_subclass: " + str(lamost_LRS_subclass)), 
        figsize=(15, 7), xlim = xlim)
    plt.annotate("H9 3835", xy=(3835, df_lamost_LRS_CSV_4600['Flux'].median()), rotation=-90)
    plt.annotate("H8/He I 3889", xy=(3889, df_lamost_LRS_CSV_4600['Flux'].median()), rotation=-90)
    plt.annotate("Ca II K 3934", xy=(3934, df_lamost_LRS_CSV_4600['Flux'].median()), rotation=-90)
    plt.annotate("H Epsilon 3970", xy=(3970, df_lamost_LRS_CSV_4600['Flux'].median()), rotation=-90)
    plt.annotate("H Delta 4101", xy=(4101, df_lamost_LRS_CSV_4600['Flux'].median()), rotation=-90)
    plt.annotate("H Gamma 4340", xy=(4340, df_lamost_LRS_CSV_4600['Flux'].median()), rotation=-90)
    plt.annotate("He I 4471", xy=(4471, df_lamost_LRS_CSV_4600['Flux'].median()), rotation=-90)

    # ====================================================================================
    xlim2 = (6000, 9000)
    df_lamost_LRS_CSV_6000 = df_lamost_LRS_CSV[df_lamost_LRS_CSV['Wavelength'] >= 6000]
    df_lamost_LRS_CSV_6000.plot(x='Wavelength',y='Flux', linestyle='--', color='r', 
        title = ('Gaia Source ID: ' + str(source_id_target)  + '\nlamost_LRS_obsid: ' + str(lamost_LRS_obsid) + "\nlamost_LRS_subclass: " + str(lamost_LRS_subclass)), 
        figsize=(15, 7), xlim = xlim2)
    plt.annotate("H Alpha 6500", xy=(6500, df_lamost_LRS_CSV_6000['Flux'].median()), rotation=-90)
    plt.annotate("Ca 8500", xy=(8500, df_lamost_LRS_CSV_6000['Flux'].median()), rotation=-90)
    plt.annotate("Ca 8544", xy=(8544, df_lamost_LRS_CSV_6000['Flux'].median()), rotation=-90)
    plt.annotate("Ca 8664", xy=(8664, df_lamost_LRS_CSV_6000['Flux'].median()), rotation=-90)

    plt.show()

print("\nDone!")
```
